## 1 Від'ємний біноміальний розподіл

Більш детальна інформація в [документації](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.nbinom.html).

```python
from scipy.stats import nbinom
```

```python
n = 5
p = 0.4
```

### 1.1 Щільність

```python
x = np.arange(-0.5, n*5+1, 0.5)
```

Виведіть значення функції щільності від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
nbinom_pmf = nbinom.pmf(x, n, p)
nbinom_pmf
```

Зобразіть графічно значення функції щільності від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
plt.stem(nbinom_pmf)
plt.show()
```

Зобразіть графічно значення функції щільності від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x1$.

```python
x1 = np.arange(0, n*5+1)
```

```python
plt.stem(nbinom.pmf(x1, n, p))
plt.show()
```

### 1.2 Функція розподілу

Виведіть значення функції розподілу від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
nbinom_cdf = nbinom.cdf(x, n, p)
nbinom_cdf
```

Зобразіть графічно значення функції розподілу від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
plt.plot(x, nbinom_cdf)
plt.show()
```

### 1.3 Випадкові величини

Згенеруйте $100$ випадкових величин з від'ємного біноміального розподілу з параметрами $n, p$.

```python
nbinom_rvs = nbinom.rvs(n, p, size=100)
nbinom_rvs
```

Виведіть кількість кожного елемента в отриманому випадковому векторі.

```python
for i in set(nbinom_rvs):
    print('Elem {} - {} times'.format(i, list(nbinom_rvs).count(i)))
```

Зобразіть графічно кількість кожного елемента в отриманому випадковому векторі.

```python
plt.hist(nbinom_rvs)
plt.show()
```

### 3.1 Біноміальний розподіл

Зобразіть графічно функцію щільності, функцію розподілу та гістограму $10000$ випадкових величин біноміального розподілу з параметрами:
1. $n = 15, p = 0.8$
2. $n = 15, p = 0.2$
3. $n = 20, p = 0.4$
4. $n = 20, p = 0.6$
5. $n = 7, p = 0.5$
6. $n = 25, p = 0.5$
7. $n = 100, p = 0.5$

```python
from scipy.stats import binom
```

```python
def discr_rv_plot(*args, distribution=None, n=15):
    discr_rv = distribution(*args)
    x = np.arange(0, n+1)
    fig, axes = plt.subplots(3, figsize=(7, 12))
    axes[0].stem(discr_rv.pmf(x))
    axes[0].set_title('Probability mass function')
    axes[1].plot(x, discr_rv.cdf(x))
    axes[1].set_title('Cumulative distribution function')
    axes[2].hist(discr_rv.rvs(size=10000))
    axes[2].set_title('Random variables')
    plt.show()
```

```python
for n, p in [(15, 0.8), (15, 0.2), (20, 0.4), (20, 0.6), (7, 0.5), (20, 0.5), (100, 0.5)]:
    print('For n = {}, p = {}:'.format(n, p))
    discr_rv_plot(n, p, distribution=binom, n=n)
```

## 4 Апроксимація біноміального розподілу розподілом Пуассона

Покажіть схожість біноміального розподілу та розподілу Пуассона (їх функцій щільності) з відповідними параметрами.
Наприклад:
- $n = 100, p = 0.05$
- $n = 100, p = 0.1$
- $n = 100, p = 0.15$
- $n = 100, p = 0.2$
- $n = 100, p = 0.35$
- $n = 100, p = 0.5$
- $n = 100, p = 0.75$

```python
def approx(n, p):
    binom_d = binom(n, p)
    poisson_d = poisson(p*n)
    x = np.arange(0, n+1)
    fig, axes = plt.subplots(3, 2, figsize=(10, 15))
    
    axes[0, 0].stem(x, binom_d.pmf(x))
    axes[0, 0].set_title('Binom probability mass function')
    axes[0, 1].stem(x, poisson_d.pmf(x))
    axes[0, 1].set_title('Poisson probability mass function')

    axes[1, 0].plot(x, binom_d.cdf(x))
    axes[1, 0].set_title('Binom cumulative distribution function')
    axes[1, 1].plot(x, poisson_d.cdf(x))
    axes[1, 1].set_title('Poisson cumulative distribution function')
    
    axes[2, 0].hist(binom_d.rvs(size=10000))
    axes[2, 0].set_title('Binom random variables')
    axes[2, 1].hist(poisson_d.rvs(size=10000))
    axes[2, 1].set_title('Poisson random variables')


    plt.show()
```

```python
for p in [0.05, 0.1, 0.15, 0.2, 0.35, 0.5, 0.75]:
    print('For n = 100, p =', p)
    approx(100, p)
```

Висновок: зі збільшенням ймовірності успіху $p$ збільшується розбіжність розподілів.
